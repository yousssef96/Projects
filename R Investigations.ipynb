### (vi)

We know that if we have two linear regression models, M1 and M2, say, are nested if the explanatory variables in M1 are a subset of the explanatory variables in M2.  

Since the explanatory variable `year` from Model A is also included in Models B, C, and D, we can conclude that Model A is nested within each of these models.

```r
#Model A 
modelA <- lm(nestNumber ~ year, data = blackTerns)

#Model B
modelB <- lm(nestNumber ~ year + subcolony, data = blackTerns)

#Model C 
modelC <- lm(nestNumber ~ factor(blackTerns$year),  data = blackTerns)

#Model D
modelD <- lm(nestNumber ~ factor(blackTerns$year) + subcolony,  data = blackTerns)
```

```r
# Test whether subcolony should be included in the model
anova(modelA, modelB)

# Test whether year should be treated as a factor in the model
anova(modelA, modelC)

# Test whether year should be treated as a factor and for subcolony to be in the model
anova(modelA, modelD)
```

Even though the residual sum of squares (RSS) for Model A is larger than other models. The $p$-values of these tests are quite large. Therefore, there is no evidence that `subcolony` should be added to the model , nor that `year` should be treated as a factor rather than a numeric variable.  

Following the principle of parsimony, which suggests to include in the model those explanatory variables which significantly
increase the fit of the model, we can conclude that Model A is better than each of Models B, C and D

A stepwise regression procedure could be used to help with choosing

```r
#Full model with all the variables
fullModel <- lm(nestNumber ~ lakeLevel + area + areaHatchSuccess + areaHab +  areaUnhab, data=blackTerns)


#Null model with no variables
nullModel <- lm(nestNumber ~ 1, data=blackTerns)
```

### (i)

The the number of nests in that subcolony  is a count response variable that can only take non-negative integer
values. Linear regression would include negative values as possible
outcomes of the response and would also include decimal values. Both
negative values and decimals are not possible for count (integer) data,
and so linear regression may not be appropriate.  

The Poisson distribution
is discrete, to match the discrete data which is appropriate for a count response variable. In addition, the probability of
a negative value for `nestNumber` is zero under the Poisson distribution


### (ii)

I applied stepwise procedure using AIC  to identify the best-fitting model. The stepwise procedure started with the full model and then iteratively removed explanatory variables that did not contribute significantly, based on AIC values, until the model with the lowest AIC was found. 

```r
# Fit the null model
nullModel <- glm(nestNumber ~ 1, family = poisson, data = blackTerns)

# Fit the full model
fullModel <- glm(nestNumber ~  lakeLevel + area + areaHatchSuccess + areaHab + areaUnhab
                 , family = poisson, 
                 data = blackTerns)
```

Running the code cell below to run the step() function in both directions starting with the full model.

```r
step(fullModel, direction = "both", 
     scope = list(lower = nullModel, upper = fullModel))
```

Starting with the full model, the chosen model is:

nestNumber ~ lakeLevel + area + areaHatchSuccess + areaHab.

So, the selected model includes all of the possible explanatory variables except for `areaUnhab`.

Running the code cell below to run the step() function in both directions, but this time starting with the null model.

```r
step(nullModel, direction = "both", 
     scope = list(lower = nullModel, upper = fullModel))
```

The model selected when starting with the null model is the same as the model selected when we started from the full model. That is, the selected model is:

nestNumber ~ lakeLevel + area + areaHatchSuccess + areaHab.

Sensitivity to Outliers: MSE squares the residuals (the differences between predicted and actual values), which means that larger errors have a disproportionately high impact on the overall score. This can skew the evaluation if there are outliers in the data, making the model appear worse than it might be for the majority of the data points. In this case, we already know that there are three outliers in the dataset, which could be inflating the MSE values.  


MSE assumes normally distributed errors. When this assumption is violated—such as when the error distribution is skewed or has heavy tails—MSE may not be a reliable measure of model performance.
```r
anaemiaPC <- prcomp(anaemia, rank = 2)


anaemiaDiss <- dist(anaemia)
ClusterSol <- hclust(anaemiaDiss)


clusterAllocation <- cutree(ClusterSol, k = 3)
plot(anaemiaPC$x, col = clusterAllocation, main = "PCA of Anaemia Data by Cluster",  las=1)
legend("bottomleft",  title = "Clusters", legend = 1:3, col = 1:3, pch = 1)
```

```r
table(cutree(ClusterSol, 3))
```

There are 59 observations in cluster 1, 93 in cluster 2 and 38 in cluster 3.

```r
plot(ClusterSol)
```

Based on the dendrogram the biggest changes in dissimilarity appear to happen going from 2 to 3 clusters at about height 60–70. So it does seem reasonable to split the data into three clusters.

### (c)

I will use the dissimilarity matrix along with silhouette statistics to help justify why a three-cluster solution may be the most appropriate choice

```r
dissimPlot(anaemiaDiss, clusterAllocation)
```

On the plot, the blocks along the main diagonal stand out as much lighter than the other blocks. This indicates that observations within each cluster are much closer than observations in different clusters. So, the clustering has been generally successful in identifying observations that are similar.

```r
anaemiaSilhouette <- silhouette(anaemia, clusterAllocation)
silhouettePlot(anaemiaSilhouette)
```

For all three clusters, the values of the silhouette statistics are largely positive, suggesting that the clusters are reasonable ones. However, in Cluster 2, some observations have negative silhouette values, suggesting that these points are closer to a different cluster than the one they were assigned to.

Based on the two plot above, we can conclude that the three-cluster solution does look reasonable

### (d)

K-means clustering tends to converge to a stable solution, meaning that after a certain number of iterations, the results no longer change. Although the algorithm doesn’t guarantee how quickly convergence will occur, it is generally efficient and well-suited to handling large datasets. This makes it a practical choice for assigning similar data points when the data are broken down into many smaller geographic regions. Additionally, once the cluster centers have been identified, new data points can be easily assigned to the appropriate cluster based on their distance to these centers.
